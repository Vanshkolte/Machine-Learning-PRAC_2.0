## Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler

stand = StandardScaler()
stand.fit(data[['balance']])
```

```python
data["balance_stand"] = pd.DataFrame(stand.transform(data[['balance']]),columns =["x"])
```

```python
data.head(2)
```

```python
## # separating the data and label
```

```python
X = data.drop(columns=['default'],axis=1)
Y = data['default']
```

```python
print(X)
print(Y)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
Rf = RandomForestClassifier()
```

```python
Rf.fit(X_train ,Y_train)
```

```python
## test perdict
pre = Rf.predict(X_test)
pre
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, pre)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(Y_test,pre))
```

```python
## Hypertuning
```

```python
# Define hyperparameter grid
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [2, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt'],
    'bootstrap': [True, False]
}

# Setup GridSearchCV
grid_search = GridSearchCV(estimator=Rf, param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2)

# Fit the model
grid_search.fit(X_train, Y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Evaluate on the test set
best_rf = grid_search.best_estimator_

y_pred = best_rf.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, y_pred)

print("Test Accuracy:", accuracy)
```

## ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input 
from livelossplot import PlotLossesKerasTF
```

```python
sq = Sequential()

sq.add(Input(shape = (13,), name = "InputLayer"))

sq.add(Dense(14, activation = "relu" , name = "H1"))
       
sq.add(Dense(16, activation = "relu" , name = "H2"))

sq.add(Dense(1,activation = "sigmoid" , name = "OutputLayer"))

sq.compile(loss = "binary_crossentropy", optimizer = "adam" , metrics = ["accuracy"])
       
```

```python
sq.fit(X_train,Y_train,epochs =20,batch_size = 32,validation_data =(X_test , Y_test),callbacks = [PlotLossesKerasTF()])

# Evaluate the model
loss, accuracy = sq.evaluate(X_test, Y_test)
print("ANN Model Accuracy:", accuracy)
```

```python
## HyperTuning
sq = Sequential()

sq.add(Input(shape = (13,), name = "InputLayer"))

sq.add(Dense(14, activation = "relu" , name = "H1"))

sq.add(Dense(16, activation = "relu" , name = "H2"))

sq.add(Dense(18, activation = "relu" , name = "H3"))
       
sq.add(Dense(20, activation = "relu" , name = "H4"))

sq.add(Dense(1,activation = "sigmoid" , name = "OutputLayer"))

sq.compile(loss = "binary_crossentropy", optimizer = "adam" , metrics = ["accuracy"])

sq.fit(X_train,Y_train,epochs =20,batch_size = 32,validation_data =(X_test , Y_test),callbacks = [PlotLossesKerasTF()])

# Evaluate the model
loss, accuracy = sq.evaluate(X_test, Y_test)
print("ANN Model Accuracy:", accuracy)
```
