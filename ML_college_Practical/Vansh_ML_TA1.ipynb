# Univariate Analysis: Histograms

```python
# Univariate Analysis: Histograms
import seaborn as sns
import matplotlib.pyplot as plt

# Univariate analysis of numerical variables
def univariate_analysis_numerical(df, column):
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Univariate analysis of categorical variables
def univariate_analysis_categorical(df, column):
    sns.countplot(data=df, x=column)
    plt.title(f'Count of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()

# Perform univariate analysis on numerical variables
univariate_analysis_numerical(dt, 'age')
print("The histogram for 'age' will reveal the age distribution of the employees. A normal distribution would imply a balanced age spread, while a skew suggests age concentration in certain ranges.")

univariate_analysis_numerical(dt, 'salary')
print("The histogram for 'salary' will show the distribution of salaries within the company. If the distribution is right-skewed, it suggests that most employees earn lower salaries with fewer earning higher salaries.")

univariate_analysis_numerical(dt, 'annual_bonus')
print("The histogram for 'annual_bonus' will show how bonuses are distributed. A right-skewed distribution suggests that most bonuses are on the lower end, with fewer large bonuses.")

# Perform univariate analysis on categorical variables
univariate_analysis_categorical(dt, 'department')
print("The count plot for 'department' will reveal the number of employees in each department, showing which departments are larger or smaller in the company.")

univariate_analysis_categorical(dt, 'company')
print("The count plot for 'company' will show the distribution of employees across different companies, indicating if one company has more employees in the dataset than others.")
```

## Bivariate : boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bivariate analysis between two numerical variables
def bivariate_analysis_numerical(df, column1, column2):
    sns.boxplot(data=df, x=column1, y=column2)
    plt.title(f'Scatter Plot of {column1} vs {column2}')
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.show()

# Bivariate analysis between a numerical and a categorical variable
def bivariate_analysis_categorical_numerical(df, categorical_column, numerical_column):
    sns.boxplot(data=df, x=categorical_column, y=numerical_column)
    plt.title(f'Boxplot of {numerical_column} by {categorical_column}')
    plt.xlabel(categorical_column)
    plt.ylabel(numerical_column)
    plt.show()

# Perform bivariate analysis on two numerical variables
bivariate_analysis_numerical(dt, 'age', 'salary')
print("The scatter plot for 'age' vs 'salary' will reveal if there's any relationship between age and salary, such as whether older employees tend to earn more.")

bivariate_analysis_numerical(dt, 'years_in_the_company', 'salary')
print("The scatter plot for 'years_in_the_company' vs 'salary' will show if salary increases with tenure in the company.")

# Perform bivariate analysis on a categorical and a numerical variable

bivariate_analysis_categorical_numerical(dt, 'company', 'annual_bonus')
print("The boxplot for 'annual_bonus' by 'company' will show how bonuses vary between different companies.")
```

## Multivariate 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function for multivariate analysis with three numerical variables using pairplot
def multivariate_analysis_numerical(df, columns):
    sns.pairplot(df[columns])
    plt.suptitle('Pairplot for Multivariate Analysis', y=1.02)
    plt.show()

# Function for multivariate analysis with two numerical and one categorical variable using a scatter plot with hue
def multivariate_analysis_categorical_numerical(df, categorical_column, numerical_column1, numerical_column2):
    sns.scatterplot(data=df, x=numerical_column1, y=numerical_column2, hue=categorical_column)
    plt.title(f'Scatter Plot of {numerical_column2} vs {numerical_column1} by {categorical_column}')
    plt.xlabel(numerical_column1)
    plt.ylabel(numerical_column2)
    plt.show()

# Example: Perform multivariate analysis with three numerical variables
multivariate_analysis_numerical(dt, ['salary', 'years_in_the_company', 'prior_years_experience'])

# Conclusion for the multivariate pairplot
print("The pairplot reveals the relationships between 'salary', 'years_in_the_company', and 'prior_years_experience'. It might show that 'years_in_the_company' has a stronger positive correlation with 'salary'.")

# Example: Perform multivariate analysis with a categorical and two numerical variables
multivariate_analysis_categorical_numerical(dt, 'department', 'salary', 'annual_bonus')

# Conclusion for the multivariate scatter plot
print("The scatter plot with hue for 'department' reveals how the relationship between 'salary' and 'annual_bonus' changes based on the department. It might show that different departments have different salary and bonus trends.")

# Example: Perform multivariate analysis with another categorical and two numerical variables
multivariate_analysis_categorical_numerical(dt, 'company', 'salary', 'years_in_the_company')

# Conclusion for the multivariate scatter plot
print("The scatter plot with hue for 'company' shows the impact of company type on the relationship between 'salary' and 'years_in_the_company'. It might reveal that employees in certain companies receive higher salaries after spending more years in the company.")
```
